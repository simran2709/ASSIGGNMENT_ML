QUES1. What is Simple Linear Regression?
-  It finds the best straight line that predicts Y from X.

QUES2.What are the key assumptions of Simple Linear Regression?
- Linearity  
- Constant variance (homoscedasticity)  
- Independence of errors  
- Normal distribution of errors  
- No big outliers

QUES3.What is heteroscedasticity, and why is it important to address in regression models?
- It's when error spread is unequal. It leads to biased estimates and unreliable predictions.

QUES4.What is Multiple Linear Regression?
- It predicts output using more than one input (e.g. X1, X2…).

QUES5. What is polynomial regression, and how does it differ from linear regression?
- Polynomial regression fits curves (e.g. X²), while linear regression fits straight lines.

QUES10.You detect heteroscedasticity and multicollinearity in your regression model.How do you fix and make the model robust?
- Use log or square root to fix heteroscedasticity  
- Remove or combine similar features (to reduce multicollinearity)  
- Use Ridge or Lasso regression  
- Cross-validate the model



```python
#QUES6.Implement a Python program to fit a Simple Linear Regression model to:
#X = [1, 2, 3, 4, 5], Y = [2.1, 4.3, 6.1, 7.9, 10.2]
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
import numpy as np

X = np.array([1,2,3,4,5]).reshape(-1,1)
Y = np.array([2.1,4.3,6.1,7.9,10.2])
model = LinearRegression().fit(X, Y)
plt.scatter(X,Y); plt.plot(X, model.predict(X), color='r'); plt.show()
```

```python
#QUES7.Fit a Multiple Linear Regression model on given data & check for multicollinearity using VIF.
import pandas as pd
from sklearn.linear_model import LinearRegression
from statsmodels.stats.outliers_influence import variance_inflation_factor

df = pd.DataFrame({'Area':[1200,1500,1800,2000],'Rooms':[2,3,3,4],'Price':[250000,300000,320000,370000]})
X = df[['Area','Rooms']]
X['Intercept'] = 1
vif = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
print(vif)
```

```python
#QUES8.Implement polynomial regression (degree 2) on:
#X = [1, 2, 3, 4, 5], Y = [2.2, 4.8, 7.5, 11.2, 14.7]
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
import numpy as np, matplotlib.pyplot as plt

X = np.array([1,2,3,4,5]).reshape(-1,1)
Y = np.array([2.2,4.8,7.5,11.2,14.7])
X_poly = PolynomialFeatures(degree=2).fit_transform(X)
model = LinearRegression().fit(X_poly, Y)
plt.scatter(X,Y); plt.plot(X, model.predict(X_poly), color='r'); plt.show()
```

```python
#QUES9.Create a residuals plot for:
#X = [10, 20, 30, 40, 50], Y = [15, 35, 40, 50, 65]
from sklearn.linear_model import LinearRegression
import numpy as np, matplotlib.pyplot as plt

X = np.array([10,20,30,40,50]).reshape(-1,1)
Y = np.array([15,35,40,50,65])
model = LinearRegression().fit(X,Y)
res = Y - model.predict(X)
plt.scatter(model.predict(X), res); plt.axhline(0, color='r'); plt.show()
```
